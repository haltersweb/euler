## first try (not final solution)

```python
# FIRST TRY (NOT FINAL SOLUTION)

# make multiples
# test for palindrome
# append palindrome
palindrome = []
largest_palindrome = 0
for j in reversed(range(100,1000)):
    for i in reversed(range(j,1000)):
        product = i*j
        if product == int(str(product)[::-1]) and product not in palindrome:
            palindrome.append(product)
            if product > largest_palindrome:
                largest_palindrome = product
print('largest palindrome is:', largest_palindrome)
print(len(palindrome))
palindrome
```

## using `continue` in i loop

Continue to next `i` without running the palindrome test, when product becomes less than the current largest palindrome)

```python
import time
start = time.time()

palindrome = []
largest_palindrome = 0
for j in reversed(range(100,1000)):
    for i in reversed(range(j,1000)):
        product = i*j
        if product < largest_palindrome:
            continue
        if product == int(str(product)[::-1]) and product not in palindrome:
            palindrome.append(product)
            largest_palindrome = product
print('largest palindrome is:', largest_palindrome)

end = time.time()
print((end-start) * 10**3, "ms")
print(len(palindrome))
palindrome
```

## using `break` in i loop

Stop `i` loop completely and continue to next `j`, when product becomes less than the current largest palindrome.

```python
import time
start = time.time()

palindrome = []
largest_palindrome = 0
for j in reversed(range(100,1000)):
    for i in reversed(range(j,1000)):
        product = i*j
        if product < largest_palindrome:
            break
        if product == int(str(product)[::-1]) and product not in palindrome:
            palindrome.append(product)
            largest_palindrome = product
print('largest palindrome is:', largest_palindrome)

end = time.time()
print((end-start) * 10**3, "ms")
print(len(palindrome))
palindrome
```

## other people's solutions

### example 1

The palindrome can be written as:

abccba

Which then simpifies to:

100000a + 10000b + 1000c + 100c + 10b + a

And then:

100001a + 10010b + 1100c

Factoring out 11, you get:

11(9091a + 910b + 100c)

So the palindrome must be divisible by 11.  Seeing as 11 is prime, at least one of the numbers must be divisible by 11.  So brute force in Python, only with less numbers to be checked:

```python
def c():
	max = maxI = maxJ = 0
	i = 999
	j = 990
	while (i > 100):
		j = 990
		while (j > 100):
			product = i * j
			if (product > max):
				productString = str(product)
				if (productString == productString[::-1]):
					max = product
					maxI = i
					maxJ = j
			j -= 11
		i -= 1
	return max, maxI, maxJ

import time
start = time.time()
print(c())
end = time.time()
print((end-start) * 10**3, "ms")
```

## final solution

```python
import time
start = time.time()



largest_palindrome = (0, 0, 0) # (i, j, product)
# start with 999
for j in reversed(range(100,1000)):
    # use j as start of range so as not to duplicate product (955 x 100 == 100 x 995)
    for i in reversed(range(j,1000)):
        # first product will be 999 x 999
        product = i*j
        # don't do unnecessary product calculations if they will be less than the current largest_palindrome
        if product < largest_palindrome[-1]:
            break
        # product palendrome test
        if product == int(str(product)[::-1]):
            largest_palindrome = (i, j, product)
print('largest palindrome is: {} x {} = {}'.format(*largest_palindrome))


end = time.time()
print((end-start) * 10**3, "ms")
```
